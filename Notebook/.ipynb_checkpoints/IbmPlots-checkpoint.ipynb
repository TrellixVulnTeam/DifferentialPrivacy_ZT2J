```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
dataibm =pd.read_excel(open('analysistest.xlsx', 'rb'), sheet_name='IBM') 
dataibm.columns =[column.replace(" ", "_") for column in dataibm.columns]
dataibm.head(1)
```

```python
fig, axs = plt.subplots(ncols=2,figsize=(15, 6))
sns.lineplot(x="epsilon(0_to_100_log_step)", y="accuracy(0_to_100_log_step)",  data=dataibm,ax=axs[0]);
sns.lineplot(x="epsilon(0_to_10_linear)", y="accuracy(0_to_10_linear)",  data=dataibm,ax=axs[1])
```
